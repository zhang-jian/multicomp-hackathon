```python
%matplotlib inline

import pickle

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.externals import joblib

sns.set_context("poster")
```

```python
NUM_SECONDS = 3
NUM_SAMPLES = 5
```

```python
with open("features/{}s_{}_img_a_features.pkl".format(NUM_SECONDS, NUM_SAMPLES), "rb") as f:
    style_features, y = pickle.load(f, encoding="latin1")
```

```python
tsne = TSNE(n_components=2, verbose=1)
```

```python
style_tsne = tsne.fit_transform(style_features)
```

```python
colors = sns.xkcd_palette(["bright red", "dark red", "orange", "yellow", "bright green", "dark green", "bright blue", "dark blue", "bright purple", "purple"])
genres = ["blues", "classical", "country", "disco", "hiphop", "jazz", "metal", "pop", "reggae", "rock"]
recs = []
# plt.figure(figsize=(8, 8))
for i, genre in enumerate(genres):
    idxs = [j for j in range(len(y)) if y[j] == i]
    plt.scatter(style_tsne[idxs, 0], style_tsne[idxs, 1], c=colors[i])
    recs.append(mpatches.Rectangle((0, 0), 1, 1, color=colors[i]))
# plt.xlim(-40, 40)
# plt.ylim(-40, 40)
_ = plt.legend(recs, genres, loc=4)#, fontsize="xx-large")
plt.savefig("tsne.jpg", dpi=200)
```
